```python
categori_choco['transaction_date'] = pd.to_datetime(categori_choco['transaction_date'])

# Filter data dari bulan Januari hingga Juni
bulanjanjun= categori_choco[(categori_choco['transaction_date'].dt.month >= 1) & (categori_choco['transaction_date'].dt.month <= 6)]

# Kelompokkan data berdasarkan bulan dan hitung total revenue
monthly_revenue = bulanjanjun.groupby(bulanjanjun['transaction_date'].dt.month)['unit_price'].sum()
```

```python
from sklearn.linear_model import LinearRegression

# Data untuk bulan Juli hingga Desember
months_to_forecast = list(range(7, 13))

# Persiapan data dari bulan Januari hingga Juni
train_data = bulanjanjun.groupby(bulanjanjun['transaction_date'].dt.month)['unit_price'].sum()
X_train = np.array(train_data.index).reshape(-1, 1)  # Bulan Januari hingga Juni
y_train = train_data.values
# Buat model regresi linier
model = LinearRegression()
model.fit(X_train, y_train)

# Lakukan forecast untuk bulan Juli hingga Desember
X_forecast = np.array(months_to_forecast).reshape(-1, 1)  # Bulan Juli hingga Desember
forecast = model.predict(X_forecast)

# Menampilkan hasil prediksi
print("Perkiraan Pendapatan untuk Bulan Juli hingga Desember:")
for month, revenue in zip(months_to_forecast, forecast):
    print(f"Bulan {month}: {revenue:.2f}")

# Buat visualisasi time series dengan perkiraan
plt.figure(figsize=(10, 6))
plt.plot(train_data.index, train_data, marker='o', linestyle='-', label='Historical Revenue (Jan-Jun)')
plt.plot(months_to_forecast, forecast, marker='o', linestyle='-', label='Forecast (Jul-Dec)', color='tan')
plt.title('Forecasting Categori Drinking chocolate Revenue (Linear Regression)')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.grid(True)



# Menambahkan label untuk nilai perkiraan
for month, revenue in zip(months_to_forecast, forecast):
    plt.text(month, revenue, f'{revenue:.2f}', ha='center', va='bottom', fontsize=10, color='black', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.2'))

plt.tight_layout()
plt.show()

```
